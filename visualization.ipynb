```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("healthexp")
df
```

- Line Plot - Life Expectancy Trend

```python
plt.figure(figsize=(10,5))
for c in df['Country'].unique():
    d = df[df['Country']==c]
    plt.plot(d['Year'], d['Life_Expectancy'], label=c)

plt.legend()
plt.colorbar if False else None
```

- Blue Line Plot

```python
plt.plot(df['Year'], df['Life_Expectancy'], color='blue')
```

- Red Line Plot

```python
plt.plot(df['Spending_USD'], df['Life_Expectancy'], color='red')
```

- Green Scatter Plot

```python
plt.scatter(df['Spending_USD'],
            df['Life_Expectancy'],
            color='green')
```
