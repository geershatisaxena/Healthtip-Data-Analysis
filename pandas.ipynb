- Country with highest average life expectancy.

```python
df.groupby("Country")["Life_Expectancy"].mean().idxmax()
```

- Country with lowest average life expectancy

```python
df.groupby("Country")["Life_Expectancy"].mean().idxmin()
```

- Country with highest average expenditure.

```python
df.groupby("Country")["Spending_USD"].mean().idxmax()
```

- Country with lowest average expenditure

```python
df.groupby("Country")["Spending_USD"].mean().idxmin()
```

- Find the country with the highest life expectancy in each year.

```python
df.loc[df.groupby("Year")["Life_Expectancy"].idxmax(),
       ["Year", "Country", "Life_Expectancy"]]
```

- Find the country with the lowest life expectancy in each year.

```python
df.loc[df.groupby("Year")["Life_Expectancy"].idxmin(),
       ["Year", "Country", "Life_Expectancy"]]
```

- Calculate cumulative health expenditure for each country

```python
df["Cumulative_Expenditure"] = (
    df.groupby("Country")["Spending_USD"].cumsum()
)
df["Cumulative_Expenditure"]
```

- Calculate cumulative average life expectancy for each country

```python
df["CumAvg_Life_Expectancy"] = (
    df.groupby("Country")["Life_Expectancy"]
      .expanding()
      .mean()
      .reset_index(level=0, drop=True)
)
df["CumAvg_Life_Expectancy"]
```

- Find the percentage change in health expenditure for each country

```python
df["Exp_Percent_Change"] = (
    df.groupby("Country")["Spending_USD"]
      .pct_change() * 100
)
```

- Find the year-over-year difference in life expectancy for each country

```python
df["LifeExp_Difference"] = (
    df.groupby("Country")["Life_Expectancy"]
      .diff()
)

```

- Rank countries by health expenditure within each year.

```python
df["Exp_Rank"] = (
    df.groupby("Year")["Spending_USD"]
      .rank(ascending=False)
)
```

- Find the top 3 countries by life expectancy in each year.

```python
top3 = (
    df.groupby("Year", group_keys=False)
      .apply(lambda x: x.nlargest(3, "Life_Expectancy"))
)
```

- Create a MultiIndex using Country and Year.

```python
df["Rolling_3yr_Exp"] = (
    df.groupby("Country")["Spending_USD"]
      .transform(lambda x: x.rolling(3).mean())
)
```

- Find the rolling 3-year average health expenditure for each country.

```python
df["Rolling_3yr_Exp"] = (
    df.groupby("Country")["Spending_USD"]
      .transform(lambda x: x.rolling(3).mean())
)
```
